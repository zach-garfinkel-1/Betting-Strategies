# Famous Betting Strategies and Monte Carlo Simulations

- Simple 50/50 Odds, no strategy, control case
- Martingale Strategy
- Labouchere Strategy

Do they really work?

### Imports

```python
import random
import numpy as np
import pandas as pd
from datascience import *

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

## No Strategy (Control)
- 50/50 odds of winning
- 1 to 1 payout
    - Win: win 1 dollar
    - Lose: lose 1 dollar

```python
def bet(funds, wager, number_bets):
    
    '''Plots winnings/losses against number of bets, with parameters of initial funds and value of wager'''
    
    value = funds
    wager = wager
    
    x = [0]
    y = [funds]

    for i in np.arange(1,number_bets+1):
        roll = random.choice(make_array(1,2))
        if roll == 1:
            value += wager
            x.append(i)
            y.append(value)
        else: 
            value -= wager
            x.append(i)
            y.append(value)
        
    plot = plt.plot(x,y)
    
    return plot
```

```python
def bettors(number_of_bettors, number_bets):
    '''Repeats bet() function a desired number of times'''
    for i in range(number_of_bettors):
        bet(0,1,number_bets)
```

```python
bettors(50,100)
```

## Martingale Strategy
- Double your wager every time you lose!
- If you win, return to your initial wager

- This strategy requires a large number of funds to succeed, but will produce guaranteed, though small, returns

```python
def m_bet(funds, initial_wager, number_bets):
    
    '''Plots winnings/losses against number of bets, with parameters of initial funds and value of wager'''
    
    value = funds
    
    x = [0]
    y = [funds]

    count = 0
    previous_bet = 'Win'
    
    while count <= number_bets and value > 0:
        roll = random.choice(make_array(1,2))
        if previous_bet == 'Win':
            wager = initial_wager
            if roll == 1:
                value += wager
                count +=1
                x.append(count)
                y.append(value)
                previous_bet = 'Win'
            else: 
                value -= wager
                count +=1
                x.append(count)
                y.append(value)
                previous_bet = 'Lose'
        else:
            wager = wager * 2
            if roll == 1:
                value += wager
                count +=1
                x.append(count)
                y.append(value)
                previous_bet = 'Win'
            else: 
                value -= wager
                count +=1
                x.append(count)
                y.append(value)
                previous_bet = 'Lose'
                
            
    plot = plt.plot(x,y)
    
    return plot
```

```python
def martingale_bettors(number_of_bettors, number_bets):
    '''Repeats m_bet() function a desired number of times'''
    for i in range(number_of_bettors):
        m_bet(10,1,number_bets)
```

```python
martingale_bettors(10,1000)
```

## Labouchere Strategy
- This strategy includes setting an initial "set" of values that add up to your desired profit
- You bet the sum of the first and last term in your "set"
- When you win, you remove the terms that you won
- When you lose, you add what you just lost as a single term at the end of the "set"
- The concept:
    - When you win, you take away two terms
    - When you lose, you only add one term
    - So eventually, you should "run out of terms" meaning you've made your desired profit

```python
def labouchere(funds, goal, verbose=False):
    
    '''Plots winnings/losses against number of bets, with parameters of initial funds and value of wager'''
    
    starting_funds = funds
    goal = goal
    profit = 0
    broke = False
    system = [1,1,2,2,3,3,4,4]
    assert sum(system) == goal
    if verbose:
        print(system)
    
    x = [0]
    y = [funds]
    count = 0
    
    while profit < goal and broke == False:
        roll = random.choice(make_array(1,2))
        if len(system) > 1:
            wager = system[0] + system[-1]
        else:
            wager = system[0]
        if roll == 1:
            profit+= wager
            funds += wager
            count += 1
            x.append(count)
            y.append(funds)
            if funds < 0:
                broke = True
            system.pop(0)
            if len(system) > 1:
                system.pop(-1)
                if verbose:
                    print(system)
            else:
                if verbose:
                    print(system)
        else:
            profit-= wager
            funds -= wager
            count += 1
            x.append(count)
            y.append(funds)
            if funds < 0:
                broke = True
            system.append(wager)
            if verbose:
                print(system)
            if funds < 0:
                broke == True
            else:
                broke == False
                
    plot = plt.plot(x,y)
    
    return plot
```

```python
def lab_bettors(number_of_bettors):
    '''Repeats labouchere() function a desired number of times'''
    for i in range(number_of_bettors):
        labouchere(100, 20, verbose=False)
```

```python
lab_bettors(100)
```

## Comparative Analysis
- Using a realistic scenario: 
    - Betting on Black on an American Roulette Wheel
    - 18/38 or a 47.37% chance of winning

- No strategy will be a normal distribution around 0 profit, with many winners and losers
- Martingale is extremely dependent on having a large amount of emergency funds
- Labouchere predefines the goal, so bettors will either reach that goal or go broke

### No Strategy Function

```python
def no_strat(funds, wager, number_bets, number_of_bettors):
    def no_bet(funds, wager, number_bets):
        '''Plots winnings/losses against number of bets, with parameters of initial funds and value of wager'''
        value = funds
        wager = wager
        x = [0]
        y = [funds]
        for i in np.arange(1,number_bets+1):
            roll = random.random()
            if roll < 18/38:
                value += wager
                x.append(i)
                y.append(value)
            else: 
                value -= wager
                x.append(i)
                y.append(value)
        return y[-1]

    def no_bettors(number_of_bettors, number_bets):
        '''Repeats no_bet() function a desired number of times'''
        results = []
        for i in range(number_of_bettors):
            result = no_bet(funds,wager,number_bets)
            results.append(result) 
        return results
    return no_bettors(number_of_bettors, number_bets)
```

```python
no_strat(100,1,20,10)
```

### Martingale Strategy Function

```python
def martingale_strat(funds, initial_wager, number_bets, number_of_bettors):
    def mart_bet(funds, initial_wager, number_bets):
        '''Plots winnings/losses against number of bets, with parameters of initial funds and value of wager'''
        value = funds
        x = [0]
        y = [funds]
        count = 0
        previous_bet = 'Win'
        while count <= number_bets and value > 0:
            roll = random.random()
            if previous_bet == 'Win':
                wager = initial_wager
                if roll < 18/38:
                    value += wager
                    count +=1
                    x.append(count)
                    y.append(value)
                    previous_bet = 'Win'
                else: 
                    value -= wager
                    count +=1
                    x.append(count)
                    y.append(value)
                    previous_bet = 'Lose'
            else:
                wager = wager * 2
                if roll < 18/38:
                    value += wager
                    count +=1
                    x.append(count)
                    y.append(value)
                    previous_bet = 'Win'
                else: 
                    value -= wager
                    count +=1
                    x.append(count)
                    y.append(value)
                    previous_bet = 'Lose'
        return y[-1]

    def mart_bettors(number_of_bettors, number_bets):
        '''Repeats mart_bet() function a desired number of times'''
        results = []
        for i in range(number_of_bettors):
            result = mart_bet(funds,initial_wager,number_bets)
            results.append(result)
        return results
    return mart_bettors(number_of_bettors, number_bets)
```

```python
martingale_strat(100,1,20,10)
```

### Labouchere Strategy Function

```python
def labouchere_strat(funds, goal, number_of_bettors, verbose=False):   
    def labo_bet(funds, goal, verbose=False):
        '''Plots winnings/losses against number of bets, with parameters of initial funds and value of wager'''
        starting_funds = funds
        goal = goal
        profit = 0
        broke = False
        system = [1,1,2,2,3,3,4,4]
        assert sum(system) == goal
        if verbose:
            print(system)
        x = [0]
        y = [funds]
        count = 0
        while profit < goal and broke == False:
            roll = random.random()
            if len(system) > 1:
                wager = system[0] + system[-1]
            else:
                wager = system[0]
            if roll < 18/38:
                profit+= wager
                funds += wager
                count += 1
                x.append(count)
                y.append(funds)
                if funds < 0:
                    broke = True
                system.pop(0)
                if len(system) > 1:
                    system.pop(-1)
                    if verbose:
                        print(system)
                else:
                    if verbose:
                        print(system)
                if count >= 100:
                    break
            else:
                profit-= wager
                funds -= wager
                count += 1
                x.append(count)
                y.append(funds)
                if funds < 0:
                    broke = True
                system.append(wager)
                if verbose:
                    print(system)
                if funds < 0:
                    broke == True
                else:
                    broke == False
                if count >= 100:
                    break
        return y[-1]
    
    def labo_bettors(number_of_bettors):
        '''Repeats labo_bet() function a desired number of times'''
        results = []
        for i in range(number_of_bettors):
            result = labo_bet(funds,goal,verbose=False)
            results.append(result)
        return results
    return labo_bettors(number_of_bettors)
```

```python
labouchere_strat(100, 20, 10)
```

## The Results

- Bet on Black every time
- Using 1000 bettors for each strategy
- Each bettor begins with 100 dollars

**No Strategy** 
   - On average, bettors will lose around 5% of their money
   - Only about 27% of bettors ended up making money after 100 spins on the wheel
   
**Martingale Strategy**
   - On average, bettors will lose around 12% of their money
   - But about 75% of bettors ended up making money after 50 spins on the wheel
   
**Labouchere Strategy**
   - On average, bettors will lose around 10% of their money
   - But about 85% of bettors will end up hitting their predefined goal within 100 spins
